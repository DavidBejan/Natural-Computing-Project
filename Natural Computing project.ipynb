Crossover

```
#not sure it works with variable length
def single_crossover(first, second, point):
    first_new =  "".join((first[:point], second[point:] ))
    second_new = "".join((second[:point], first[point:]))
    return first_new, second_new



def crossover(pop):
    pop = np.asarray(pop)
    double = copy.deepcopy(pop)
    for i in range (0, len(pop)):
        if np.random.rand() > crossover_thresh:
            point = np.random.randint(0, individual_length)
            second = np.random.randint(0, len(pop))
            pop[i], pop[second] = single_crossover(double[i], double[second], point)
        return pop

```

Mutation

```
def mutation(pop):
    #print("pop", len(pop))
    for i in range(0, len(pop)):
        for j in range(0, len(pop[i])):
             if np.random.rand() > mutation_thresh:
                #print(i, "before mutation", pop[i])
                seq = pop[i]
                if seq[j] is "0":
                    pop[i] = seq[:j] + "1" + seq[j + 1:]
                    #print(i, "after mutation change into 1", pop[i])
                else:
                    pop[i] = seq[:j] + "0" + seq[j + 1:]
                    #print(i, "after mutation change into 0", pop[i])
    return pop

pop =  mutation(pop)
print(pop)
```

```
def selection(pop, percentage):
    return pop[:int(len(pop)/100*percentage)]
    
pop = selection(population, 10)
print(pop)
```

Next Generation

```
def next_generation(pop, percentage, selection_method):
    
    # print("current gen", pop)
    #first thing is to calculate the fitness
    pop, fitness = selection_method(pop)
    
    dic = dict(zip(pop, fitness))
    dic_sorted = sorted(dic.items(), reverse = True, key = lambda item: item[1])
    sorted_keys = [k for k,v in dic_sorted]
    selected = selection(sorted_keys, percentage)
    cross = crossover(selected)
    next_gen = mutation(cross)
    return next_gen, dic_sorted

next_gen, dic = next_generation(population, 50, maxOne)
```

Next Generation(Tournament Selection)

```
# tournament selection
def tournament_selection_random(pop, scores, k=3):
	# first random selection
	selection_index = randint(0,len(pop))
	selection = rand.choice(len(pop), size = randint(1, k), replace = False)
	#print(selection)
	for index in selection:
		#print("index", index)
        # perform tournament)
		#print(selection_index)
		if scores[index-1] > scores[selection_index-1]:
			selection_index = index
	return pop[selection_index-1]

pop, fitness = maxOne(pop)
print(tournament_selection_random(["11", "10", "00"], [3,3,3]))
```
